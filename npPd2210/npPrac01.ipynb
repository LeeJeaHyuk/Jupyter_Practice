```python
# pr01-5
#1, 3, 5, 7, 9행의 특정 index만 출력 'title' 'genre' 'box_off_num'
# df.loc[list(range(1,9,2))] 
df.loc[list(range(1,10,2)),['title', 'genre', 'box_off_num']]
```

```python
# pr01-6
# boolean indexing을 통해 특정 데이터를 추출
# 주제가 공포인 데이터만 추출
df[df['genre']=='공포'].head(2)
# 주제가 공포이면서 'title','genre' 열만 추출
df[df['genre']=='공포'][['title','genre']].head(2)
# 주제가 공포이면서 'title','genre' ,'time' 열이면서 시간이 90이하인 데이터 추출
df[(df['genre']=='공포') & (df['time'] <= 90)][['title','genre','time']].head(3)
```

```python
# pr01-6
# 관객수가 가장 많은 영화 정보
df[df['box_off_num'] == df['box_off_num'].max()]
# 관객수가 평균 이상인 영화 정보
df[df['box_off_num'] >= df['box_off_num'].mean()].head(2)
```

```python
# pr01-7 
# 특정 열을 기준으로 데이터프레임 분할 (grouping) : 특성들을 grouping한다
# 관람가를 screening_rat 컬럼 그룹핑 .groupby
df_screenRat = df.groupby('screening_rat')
# 그룹의 개수 리턴
df_screenRat.ngroups
# 키값 확인
df_screenRat.groups.keys()
```

```python
# pr01-7 
# 그룹핑한 특정 키값의 values 리턴
df_screenRat.get_group('12세 관람가').head(2)
```

```python
# pr01-7 
# 그룹핑한 후 집계함수 사용
# 각 그룹의 개수를 구해보자
df_screenRat['genre'].count()
# 각 그룹의 관객수를 구해보자
df_screenRat['box_off_num'].sum()
# 각 그룹의 최대 관객수를 구해보자
df_screenRat['box_off_num'].max()
# df_screenRat['box_off_num'].min()
# 최대 상영 시간
# df_screenRat['time'].max()
```

```python
# pr01-7 
# GroupBy.agg(func, *args, **kwargs) 이용해서 연산
# 각 그룹에서 'time','box_off_num'의 최대값을 구해보자
df_screenRat.agg('max')[['time','box_off_num']]
df_screenRat.agg({'time': min, 'box_off_num': max}) #각각 다르게 지정할 수 있다 -> indexing을 해 주어야 할 것 같다
```
